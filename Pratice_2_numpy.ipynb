```python
a = np.arange(6).reshape(3,2)
b = np.arange(6,12).reshape(3,2)
print(np.corrcoef(a,b))
```

```python
x = np.arange(6).reshape(3,2)
y = np.arange(6,12).reshape(3,2)
cov = np.cov(x,y)
std_x = np.std(x, ddof=1)
std_y = np.std(y, ddof=1)

corr = cov[0,1] / (std_x * std_y)
print(corr)
```

```python
x = np.arange(6).reshape(3,2)
y = np.arange(6,12).reshape(3,2)
x_flat = x.flatten()
y_flat = y.flatten()

cov = np.cov(x_flat, y_flat)
corr = cov[0,1] / (np.std(x_flat, ddof=1) * np.std(y_flat, ddof=1))
print(corr)
print(np.corrcoef(x.flatten(), y.flatten()))
```
