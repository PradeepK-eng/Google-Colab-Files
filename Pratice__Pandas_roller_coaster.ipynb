# Data Preparation

```python
df.head()
```

```python
df = df[['coaster_name',
    #'Length', 'Speed',
    'Location', 'Status',
    'Opening date',
      # 'Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
     #  'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #   'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
     #  'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
     #  'Track layout', 'Fastrack available', 'Soft opening date.1',
     #  'Closing date', 'Opened', 'Replaced by', 'Website',
     #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
     #  'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
        'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    # 'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.drop(['Opening date'], axis=1)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df['year_introduced']
```

```python
pd.to_numeric(df['year_introduced'])
```

```python
df.columns
```

```python
df = df[['coaster_name','Location','Status','Opening date','Manufacturer','year_introduced', 'Type_Main',
       'opening_date_clean', 'speed_mph', 'height_ft', 'Inversions_clean',
       'Gforce_clean']].copy()
```

```python
df.columns
```

```python
df =df.rename(columns = {'coaster_name': 'Coaster_Name', 'Opening date':'Opening_Date',  'year_introduced' : 'Year_Introduced', 'opening_date_clean':'Opening_Date',
                         'speed_mph': 'Speed_mph', 'height_ft': 'Height_ft',  'Inversions_clean':  'Inversions_Clean', 'Gforce_clean':'Gforce_Clean'  })
```

```python
df.head(4)
```

```python
df['Manufacturer'].value_counts()
```

```python
df['Location'].value_counts()
```

```python
ax = df['Speed_mph'].plot(kind='hist',bins=20, title='Speed in mph')
ax.set_xlabel('Speed_mph')
plt.show()
```

```python
ax = df['Speed_mph'].plot(kind='kde', title='Speed in mph')
ax.set_xlabel('Speed_mph')
plt.show()
```

```python
ax = df.plot(kind='scatter',x='Year_Introduced', y= 'Speed_mph', title='Year Vs Speed')
ax.set_xlabel("Year")
ax.set_ylabel('Speed_mph')
plt.show()
```

```python
sns.scatterplot(x='Year_Introduced', y= 'Speed_mph', data = df, hue ='Type_Main')
```
