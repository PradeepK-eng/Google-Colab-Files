# Embedded 3D Printing – Filament Video / Image Analysis
Ready-to-run example notebook


```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

print('Libraries loaded.')
```

```python
DATA_PATH = Path('data')
VIDEO_PATH = DATA_PATH / 'videos'
FRAME_PATH = DATA_PATH / 'frames'
OUTPUT_PATH = Path('results')

FRAME_PATH.mkdir(parents=True, exist_ok=True)
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)

VIDEO_EXT = '.avi'

print('Configuration complete.')
```

```python
def extract_frames(video_file, output_dir):
    cap = cv2.VideoCapture(str(video_file))
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_name = output_dir / f'frame_{frame_count:04d}.png'
        cv2.imwrite(str(frame_name), frame)
        frame_count += 1
    cap.release()
    return frame_count
```

```python
def segment_filament(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    return binary, contours
```

```python
def measure_contour(contour):
    area = cv2.contourArea(contour)
    x, y, w, h = cv2.boundingRect(contour)
    return {
        'area_px': area,
        'width_px': w,
        'height_px': h,
        'aspect_ratio': w / h if h > 0 else 0
    }
```
